```python
# 1. To Dataframe nh d liu  cho trn.

import pandas as pd
data = {
    "Product": ["Laptop", "Phone", "Chair", "Desk", "Headphone", "Monitor", "Sofa", "Keyboard", "Mouse", "Tablet"],
    "Category": ["Electronics", "Electronics", "Furniture", "Furniture", "Accessories", "Electronics", "Furniture", "Accessories", "Accessories", "Electronics"],
    "Price": [1500, 800, 120, 300, 200, 600, 700, 150, 50, 400],
    "Date": ["2024-01-15", "2024-02-10", "2024-03-15", "2024-03-18", "2024-04-12", "2024-05-20", "2024-06-20", "2024-02-08", "2024-07-14", "2024-09-30"],
    "Quantity": [2, 5, 3, 1, 4, 2, 1, 6, 8, 3]
}

df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# 2. Hin th thng tin Dataframe va to.

print(df)
```

```python
# 3. Tnh gi tr trung bnh ca ct <Price>.

mean_price = df['Price'].mean()
print("Mean Price:", mean_price)
```

```python
# 4. Nhm d liu ct <Category> v tnh tng trung bnh ct <Quantity>.

grouped_category = df.groupby('Category')['Quantity'].mean()
print(grouped_category)
```

```python
# 5. Sp xp Dataframe theo ct <Price> gim dn
 
sorted_df = df.sort_values(by='Price', ascending=False)
print(sorted_df)
```

```python
# 6. V biu  ct cho <Price>.

import matplotlib.pyplot as plt

plt.bar(df['Product'], df['Price'])
plt.xlabel('Product')   
plt.ylabel('Price')
plt.title('Price of Products')
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. V biu  phn tn cho <Quantity> v <Price>.

plt.scatter(df['Quantity'], df['Price'])
plt.title("Quantity and Price")
plt.xlabel("Quantity")
plt.ylabel("Price")
plt.show()
```

```python
# 8. V biu  ng cho ct <Date>.

plt.plot(df['Product'], df['Date'])
plt.xticks(rotation=45)  
plt.xlabel('Product')  
plt.ylabel('Date')  
plt.title('Biu  ng cho ct Date')  
plt.grid(True)  
plt.show()  
```

```python
# 9. V biu  ct th hin tng doanh thu ca tng <Category>.

grouped_category = df.groupby('Category')['Price'].sum()
grouped_category.plot(kind='bar')
plt.title("Tng doanh thu ca tng Category")
plt.xlabel("Category")
plt.ylabel("Price")
plt.show()
```

```python
# 10. Kim tra c gi tr NaN no trong DataFrame khng.

print(df.isnull().sum())
```

```python
# 11. Tm cc gi tr trng lp trong DataFrame da trn ct 'Product' v loi b cc hng trng lp, gi li hng u tin.

duplicates = df[df.duplicated(subset=['Product'], keep=False)]
print("Cc hng trng lp:\n", duplicates)

df_unique = df.drop_duplicates(subset=['Product'], keep='first')

print("\nDataFrame sau khi loi b trng lp:")
print(df_unique)
```

```python
# 12. Lc ra cc n hng c s lng (Quantity) ln hn 5

filtered_df = df[df['Quantity'] > 5]
print(filtered_df)
```

```python
# 13. Tng doanh thu c tnh theo cng thc sau:
# Revenue = Quantity  Price
# To mt ct mi trong Dataframe vi tn <Revenue> gm tng doanh thu ca cc sn phm.

df['Revenue'] = df['Quantity'] * df['Price']
print(df)
```

```python
# 14. Thm mt ct <Sales_Level> vi iu kin:
# + < 500  ->>  Low
# + 500 --- 2000  ->> Medium
# + >2000   ->> High 

def sales_level(revenue):
    if revenue < 500:
        return 'Low'
    elif revenue >= 500 and revenue <= 2000:
        return 'Medium'
    else:
        return 'High'
    
df['Sales_Level'] = df['Revenue'].apply(sales_level)
print(df)
```

```python
# 15. V biu  ct th hin s lng n hng  tng mc <Sales_Level>.

grouped_sales_level = df.groupby('Sales_Level').size()
grouped_sales_level.plot(kind='bar')    
plt.title("S lng n hng  tng mc Sales_Level")
plt.xlabel("Sales_Level")
plt.ylabel("Quantity")
plt.show()
```

```python
# 16. To mt ct mi <Discount> vi gi tr gim gi 10% nu doanh thu (Revenue) > 1000, ngc li gim gi 5%.

def discount(revenue):
    if revenue > 1000:
        return revenue * 0.1
    else:
        return revenue * 0.05
    
df['Discount'] = df['Revenue'].apply(discount)
print(df)
```

```python
# 17. Tnh doanh thu sau gim gi <Final_Revenue> bng cng thc:
# Final_Revenue = Revenue - ( Revenue  Discount ).
# To mt ct mi trong Dataframe vi tn <Final_Revenue> gm doanh thu sau gim gi.

df['Final_Revenue'] = df['Revenue'] - df['Discount']
print(df)
```

```python
# 18. To ct <Month> ly gi tr thng t ct <Date>.

df['Month'] = df['Date'].dt.month
print(df)
```

```python
# 19. V biu  ng th hin tng doanh thu theo thng.

grouped_month = df.groupby('Month')['Revenue'].sum()
grouped_month.plot(kind='line')
plt.title("Tng doanh thu theo thng")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
# 20. Lu DataFrame vi tn <BKT_MSSV_Hoten.csv-->

df.to_csv('BKT_2274802010850_TranNhatThinh.csv', index=False)
print("Saved to BKT_2274802010850_TranNhatThinh.csv")
```
