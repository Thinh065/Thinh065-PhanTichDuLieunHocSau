Phần 1: Thống kê dữ liệu
1. Hãy sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần
2. Hãy sắp xếp dữ liệu điểm DH2 tăng dần theo nhóm giới tính
3. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1,
Q2 và Q3 của DH1 theo KT
4. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1,
Q2 và Q3 của DH1 theo KT và KV
5. Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1,
Q2 và Q3 của DH1 theo KT, KV và DT

```python
# Câu 1: Hãy sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần
import pandas as pd

# Đọc file CSV
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')

# Sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần
df_sort_DH1 = df.sort_values(by='DH1')
print(df_sort_DH1)
```

```python
# Câu 2: Hãy sắp xếp dữ liệu điểm DH2 tăng dần theo nhóm giới tính

# Sắp xếp dữ liệu điểm DH2 tăng dần theo nhóm giới tính
df_sort_DH2 = df.sort_values(by=['GT', 'DH2'])
print(df_sort_DH2)
```

```python
# Câu 3: Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, Q2 và Q3 của DH1 theo KT

# Tạo pivot-table thống kê các giá trị của DH1 theo KT
pivot_table_DH1_KT = df.pivot_table(values='DH1', index='KT', 
                                    aggfunc=['count', 'sum', 'mean', 'median', 'min', 'max', 'std', 
                                             lambda x: x.quantile(0.25),  # Q1
                                             lambda x: x.quantile(0.5),   # Q2 
                                             lambda x: x.quantile(0.75)]) # Q3
pivot_table_DH1_KT.columns = ['count', 'sum', 'mean', 'median', 'min', 'max', 'std', 'Q1', 'Q2', 'Q3']
print(pivot_table_DH1_KT)
```

```python
# Câu 4: Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, Q2 và Q3 của DH1 theo KT và KV

# Tạo pivot-table thống kê các giá trị của DH1 theo KT và KV
pivot_table_DH1_KT_KV = df.pivot_table(values='DH1', index=['KT', 'KV'], 
                                       aggfunc=['count', 'sum', 'mean', 'median', 'min', 'max', 'std', 
                                                lambda x: x.quantile(0.25),  # Q1
                                                lambda x: x.quantile(0.5),   # Q2 
                                                lambda x: x.quantile(0.75)]) # Q3
pivot_table_DH1_KT_KV.columns = ['count', 'sum', 'mean', 'median', 'min', 'max', 'std', 'Q1', 'Q2', 'Q3']
print(pivot_table_DH1_KT_KV)
```

```python
# Câu 5: Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, Q2 và Q3 của DH1 theo KT, KV và DT

# Tạo pivot-table thống kê các giá trị của DH1 theo KT, KV và DT
pivot_table_DH1_KT_KV_DT = df.pivot_table(
    values='DH1',  
    index=['KT', 'KV', 'DT'],  
    aggfunc={
        'DH1': [
            'count', 'sum', 'mean', 'median', 'min', 'max', 'std',
            lambda x: x.quantile(0.25),  # Q1
            lambda x: x.quantile(0.5),   # Q2 
            lambda x: x.quantile(0.75)   # Q3
        ]
    }
)

# Đổi tên các cột
pivot_table_DH1_KT_KV_DT.columns = [
    'count', 'sum', 'mean', 'median', 'min', 'max', 'std', 'Q1', 'Q2', 'Q3'
]

# In kết quả
print(pivot_table_DH1_KT_KV_DT)
```

```python
# Câu 1: Hãy trình bày dữ liệu biến: GT

import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')

# Lập bảng tần số và tần suất
tanso = df['GT'].value_counts()
tansuat = df['GT'].value_counts(normalize=True) * 100

# Tạo bảng tổng hợp
gt_table = pd.DataFrame({
    'Tần số': tanso,
    'Tần suất (%)': tansuat
})
print(gt_table)

# Vẽ biểu đồ tần số (cột)
plt.figure(figsize=(8, 4))
tanso.plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Biểu đồ tần số của biến GT')
plt.xlabel('Giới tính')
plt.ylabel('Tần số')
plt.xticks(rotation=0)
plt.show()

# Vẽ biểu đồ tần suất (tròn)
plt.figure(figsize=(6, 6))
tansuat.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen'])
plt.title('Biểu đồ tần suất của biến GT')
plt.ylabel('')
plt.show()
```

```python
# Câu 2: Hãy trình bày dữ liệu lần lượt các biến: US_TBM1, US_TBM2 và US_TBM3

# Thống kê cơ bản
us_tbm_stats = df[['US_TBM1', 'US_TBM2', 'US_TBM3']].describe().loc[['mean', '50%', 'min', 'max', 'std']]
us_tbm_stats.rename(index={'50%': 'median'}, inplace=True)
print(us_tbm_stats)

# Vẽ biểu đồ boxplot
plt.figure(figsize=(10, 6))
df[['US_TBM1', 'US_TBM2', 'US_TBM3']].boxplot()
plt.title('Biểu đồ boxplot của US_TBM1, US_TBM2, US_TBM3')
plt.ylabel('Điểm số')
plt.show()
```

```python
# Câu 3: Hãy trình bày dữ liệu biến DT với các học sinh là nam

# Lọc dữ liệu học sinh nam
male_students = df[df['GT'] == 'M']

# Lập bảng tần số và tần suất của biến DT
dt_tanso = male_students['DT'].value_counts()
dt_tansuat = male_students['DT'].value_counts(normalize=True) * 100

# Tạo bảng tổng hợp
dt_table = pd.DataFrame({
    'Tần số': dt_tanso,
    'Tần suất (%)': dt_tansuat
})
print(dt_table)

# Vẽ biểu đồ tần số (cột)
plt.figure(figsize=(10, 6))
dt_tanso.plot(kind='bar', color='skyblue')
plt.title('Biểu đồ tần số của biến DT (học sinh nam)')
plt.xlabel('Dân tộc')
plt.ylabel('Tần số')
plt.xticks(rotation=45)
plt.show()
```

```python
# Câu 4: Hãy trình bày dữ liệu biến KV với các học sinh là nam thuộc dân tộc Kinh, có điểm thỏa mãn điều kiện (DH1 >= 5.0 và DH2 >= 4.0 và DH3 >= 4.0

# Lọc dữ liệu
loc_data = df[
    (df['GT'] == 'M') & 
    (df['DT'] == 0.0 ) & 
    (df['DH1'] >= 5.0) & 
    (df['DH2'] >= 4.0) & 
    (df['DH3'] >= 4.0)
]

# Lập bảng tần số và tần suất của biến KV
kv_tanso = loc_data['KV'].value_counts()
kv_tansuat = loc_data['KV'].value_counts(normalize=True) * 100

# Tạo bảng tổng hợp
kv_table = pd.DataFrame({
    'Tần số': kv_tanso,
    'Tần suất (%)': kv_tansuat
})
print(kv_table)

# Vẽ biểu đồ tần số (cột)
plt.figure(figsize=(8, 4))
kv_tanso.plot(kind='bar', color='lightgreen')
plt.title('Biểu đồ tần số của biến KV (học sinh nam, dân tộc Kinh, điểm thỏa mãn điều kiện)')
plt.xlabel('Khu vực')
plt.ylabel('Tần số')
plt.xticks(rotation=0)
plt.show()

```

```python
# Câu 5: Hãy trình bày dữ liệu lần lượt các biến DH1, DH2, DH3 lớn hơn bằng 5.0 và thuộc khu vực 2NT

# Lọc dữ liệu
loc_data = df[
    (df['DH1'] >= 5.0) & 
    (df['DH2'] >= 5.0) & 
    (df['DH3'] >= 5.0) & 
    (df['KV'] == '2NT')
]

# Thống kê cơ bản
dh_stats = loc_data[['DH1', 'DH2', 'DH3']].describe().loc[['mean', '50%', 'min', 'max', 'std']]
dh_stats.rename(index={'50%': 'median'}, inplace=True)
print(dh_stats)

# Vẽ biểu đồ boxplot để so sánh phân phối
plt.figure(figsize=(10, 6))
loc_data[['DH1', 'DH2', 'DH3']].boxplot()
plt.title('Biểu đồ boxplot của DH1, DH2, DH3 (điểm >= 5.0 và khu vực 2NT)')
plt.ylabel('Điểm số')
plt.show()
```

```python
# Câu 1: Trực quan dữ liệu học sinh nữ trên các nhóm XL1, XL2, XL3 dạng unstacked

import pandas as pd
import matplotlib.pyplot as plt

# Đọc file CSV
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')

# Lọc dữ liệu học sinh nữ
female_hs = df[df['GT'] == 'F']

# Tạo bảng thống kê số lượng học sinh nữ theo XL1, XL2, XL3
xl_counts = female_hs[['XL1', 'XL2', 'XL3']].apply(pd.Series.value_counts)

# Vẽ biểu đồ cột unstacked
xl_counts.plot(kind='bar', figsize=(12, 6), color=['skyblue', 'lightgreen', 'orange'])
plt.title('Số lượng học sinh nữ theo xếp loại (XL1, XL2, XL3)')
plt.xlabel('Xếp loại')
plt.ylabel('Số lượng học sinh')
plt.xticks(rotation=0)
plt.legend(title='Nhóm xếp loại')
plt.show()
```

```python
# Câu 2: Trực quan dữ liệu KQXT trên nhóm học sinh có khối thi A, A1, B thuộc khu vực 1, 2

df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')
# Lọc dữ liệu học sinh có khối thi A, A1, B và thuộc khu vực 1, 2
filtered_data = df[
    (df['KT'].isin(['A', 'A1', 'B'])) & 
    (df['KV'].isin(['1', '2']))
]

# Thống kê kết quả xét tuyển (KQXT) theo khối thi
kqxt_counts = filtered_data.groupby(['KT', 'KQXT']).size().unstack()

# Vẽ biểu đồ cột
kqxt_counts.plot(kind='bar', figsize=(10, 6), stacked=True)
plt.title('Kết quả xét tuyển theo khối thi (A, A1, B) và khu vực (1, 2)')
plt.xlabel('Khối thi')
plt.ylabel('Số lượng học sinh')
plt.xticks(rotation=0)
plt.legend(title='Kết quả xét tuyển')
plt.show()
```

```python
# Câu 4: Trực quan dữ liệu số lượng thí sinh đậu, rớt trên từng nhóm khối thi

# Thống kê số lượng thí sinh đậu/rớt theo khối thi
dau_rot_counts = df.groupby(['KT', 'KQXT']).size().unstack()

# Vẽ biểu đồ cột
dau_rot_counts.plot(kind='bar', figsize=(10, 6), stacked=True)
plt.title('Số lượng thí sinh đậu/rớt theo khối thi')
plt.xlabel('Khối thi')
plt.ylabel('Số lượng thí sinh')
plt.xticks(rotation=0)
plt.legend(title='Kết quả xét tuyển')
plt.show()
```

```python
# Câu 5: Trực quan dữ liệu số lượng thí sinh đậu rớt trên từng nhóm khu vực. 

# Thống kê số lượng thí sinh đậu/rớt theo khu vực
dau_rot_kv_counts = df.groupby(['KV', 'KQXT']).size().unstack()

# Vẽ biểu đồ cột
dau_rot_kv_counts.plot(kind='bar', figsize=(10, 6), stacked=True)
plt.title('Số lượng thí sinh đậu/rớt theo khu vực')
plt.xlabel('Khu vực')
plt.ylabel('Số lượng thí sinh')
plt.xticks(rotation=0)
plt.legend(title='Kết quả xét tuyển')
plt.show()
```

    a. Từ 0 đến dưới 5 = kém (ký hiệu “k”)
    b. Từ 5 đến dưới 7 = trung bình (ký hiệu “tb”)
    c. Từ 7 đến dưới 8 = khá (ký hiệu “k”)
    d. Từ 8 trở lên = giỏi (ký hiệu “g”)
3. Lập bảng tần số cho biến phanloait1
4. Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi biến phanlopt1
5. Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến phanlopt1

```python
# Câu 1: Vẽ biểu đồ đường Simple cho biến T1

import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV
df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

# Tính tần suất của từng giá trị T1 và sắp xếp tăng dần
counts = df["T1"].value_counts().sort_index()

# Vẽ biểu đồ đường biểu diễn tần suất
plt.figure(figsize=(10, 5))

# X là các giá trị T1, Y là tần suất
plt.plot(counts.index, counts.values, marker="o", linestyle="-", color="black")

# Thiết lập tiêu đề
plt.xlabel("T1")
plt.ylabel("Count")
plt.title("Biểu đồ đường Simple cho biến T1")

plt.xticks(counts.index, [f"{x:.3f}" for x in counts.index], rotation=90)
plt.grid(True)
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```

```python
# Câu 2: Hãy tạo biến phân loại (phanlopt1) cho môn toán (T1) như sau:
        # a. Từ 0 đến dưới 5 = kém (ký hiệu “k”)  
        # b. Từ 5 đến dưới 7 = trung bình (ký hiệu “tb”) 
        # c. Từ 7 đến dưới 8 = khá (ký hiệu “k”) 
        # d. Từ 8 trở lên = giỏi (ký hiệu “g”) 

# Tạo biến phân loại phanloait1
phanloait1 = [
    (df['T1'] < 5), # Câu a
    (df['T1'] >= 5) & (df['T1'] < 7), # Câu b
    (df['T1'] >= 7) & (df['T1'] < 8), # Câu c
    (df['T1'] >= 8) # Câu d
]
categories = ['k', 'tb', 'kh', 'g']
df['phanloait1'] = pd.cut(df['T1'], bins=[0, 5, 7, 8, 10], labels=categories, right=False)

# Kiểm tra kết quả
print(df[['T1', 'phanloait1']].head())
```

```python
# Câu 3: Lập bảng tần số cho biến phanloait1 

# Lập bảng tần số
frequency = df['phanloait1'].value_counts()
print("Bảng tần số cho biến phanloait1:")
print(frequency)
```

```python
# Câu 4: Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi biến phanlopt1

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Đọc dữ liệu từ file CSV
df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

# 2. Tạo biến phân loại (phanlopt1) cho môn Toán (T1) theo cách của bạn:
#    - 0 <= T1 < 5  => "K"
#    - 5 <= T1 < 7  => "tb"
#    - 7 <= T1 < 8  => "kh
#    - 8 <= T1      => "g
conditions = [
    (df['T1'] < 5),
    (df['T1'] >= 5) & (df['T1'] < 7),
    (df['T1'] >= 7) & (df['T1'] < 8),
    (df['T1'] >= 8)
]
choices = ["K", "tb", "kh", "g"]
df["phanlopt1"] = np.select(conditions, choices, default="Unknown")

# 3. Tính tần suất cho từng cặp (T1, phanlopt1)
counts = df.groupby(["T1", "phanlopt1"]).size().unstack(fill_value=0)
counts = counts.sort_index()  # Sắp xếp T1 theo thứ tự tăng dần

# 4. Vẽ biểu đồ đường Multiple Line cho T1 được phân loại bởi phanlopt1
# Xác định thứ tự nhóm và màu sắc cho từng nhóm
cat_order = ["g", "K", "kh", "tb"]
colors = {"g": "gold", "K": "green", "kh": "blue", "tb": "purple"}

plt.figure(figsize=(10, 5))

# Vẽ mỗi nhóm phân loại thành một đường riêng
for cat in cat_order:
    if cat in counts.columns:  # Kiểm tra nhóm tồn tại trong dữ liệu
        plt.plot(
            counts.index,         # Trục X: các giá trị T1
            counts[cat],          # Trục Y: tần suất
            marker="o",
            linestyle="-",
            label=cat,
            color=colors[cat]
        )

# Thiết lập nhãn, tiêu đề và chú thích
plt.xlabel("T1")
plt.ylabel("Count")
plt.title("Biểu đồ đường Multiple Line cho T1 được phân loại bởi phanlopt1")
plt.legend(title="Phanlopt1")

# Xoay nhãn trục X hiển thị 3 chữ số thập phân để dễ nhìn
plt.xticks(counts.index, [f"{x:.3f}" for x in counts.index], rotation=90)

# Thêm đường lưới và căn chỉnh bố cục
plt.grid(True)
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```

```python
# Câu 5: Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến phanlopt1 

choices = ["k", "tb", "kh", "g"]
df["phanlopt1"] = np.select(conditions, choices, default="Unknown")

# Tính tần suất cho từng cặp (T1, phanlopt1)
counts = df.groupby(["T1", "phanlopt1"]).size().unstack(fill_value=0)
counts = counts.sort_index()  # Sắp xếp T1 theo thứ tự tăng dần

# Vẽ biểu đồ Drop-line (Stem Plot)
cat_order = ["g", "k", "kh", "tb"]
colors = {"g": "gold", "k": "green", "kh": "blue", "tb": "purple"}

plt.figure(figsize=(10, 5))

# Vẽ drop-line cho từng nhóm
for cat in cat_order:
    if cat in counts.columns:
        # Gọi stem() trả về 3 đối tượng: markerline, stemlines, baseline
        markerline, stemlines, baseline = plt.stem(
            counts.index,   # Trục X
            counts[cat],    # Trục Y
            label=cat
        )
        # Thiết lập màu sắc
        markerline.set_markerfacecolor(colors[cat])   # Màu mặt marker
        markerline.set_markeredgecolor(colors[cat])   # Màu viền marker
        stemlines.set_color("black")                  # Màu đường dọc
        baseline.set_color("black")                   # Màu baseline

# Thiết lập tiêu đề, trục và chú thích
plt.xlabel("T1")
plt.ylabel("Count")
plt.title("Biểu đồ Drop-line cho T1 được phân loại bởi phanlopt1")
plt.legend(title="PhanlopT1")

# Hiển thị nhãn T1 (xoay 90 độ) với 3 chữ số thập phân
plt.xticks(counts.index, [f"{x:.3f}" for x in counts.index], rotation=90)

# Thêm lưới và căn chỉnh bố cục
plt.grid(True)
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```

```python
# Câu 1: Hãy mô tả và khảo sát phân phối cho biến T1:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Đọc dữ liệu
df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

# Mô tả độ tập trung và phân tán
print("Mean của T1:", df['T1'].mean())
print("Median của T1:", df['T1'].median())
print("Mode của T1:", df['T1'].mode()[0])
print("Standard Deviation của T1:", df['T1'].std())
print("Variance của T1:", df['T1'].var())
print("Range của T1:", df['T1'].max() - df['T1'].min())
print("IQR của T1:", df['T1'].quantile(0.75) - df['T1'].quantile(0.25))

# Vẽ Box-Plot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['T1'])
plt.title("Box-Plot của T1")
plt.show()

# Mô tả hình dáng lệch của phân phối T1
print("Skewness của T1:", df['T1'].skew())

# Vẽ Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['T1'], kde=True)
plt.title("Histogram của T1")
plt.show()

# Vẽ QQ-Plot
plt.figure(figsize=(8, 6))
stats.probplot(df['T1'], dist="norm", plot=plt)
plt.title("QQ-Plot của T1")
plt.show()

# Nhận xét và đánh giá về phân phối của T1:
# - Dữ liệu T1 gần như tuân theo phân phối chuẩn.
# - Không có dấu hiệu mạnh của ngoại lệ hoặc lệch phân phối.
```

```python
# Câu 2: Hãy mô tả và khảo sát phân phối cho biến T1 trên từng nhóm phân lớp (phanlopT1)
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Đọc dữ liệu
df = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

# Tạo cột phân nhóm
conditions = [
    (df['T1'] < 5),
    (df['T1'] >= 5) & (df['T1'] < 7),
    (df['T1'] >= 7) & (df['T1'] < 8),
    (df['T1'] >= 8)
]
labels = ["K", "TB", "KH", "G"]
df['phanlopT1'] = np.select(conditions, labels)

# 1. **Box-Plot theo nhóm**
plt.figure(figsize=(10, 6))
sns.boxplot(x='phanlopT1', y='T1', data=df, order=labels)
plt.title("Box-Plot của T1 theo nhóm phanlopT1")
plt.xlabel("Nhóm phân loại T1")
plt.ylabel("Giá trị T1")
plt.show()

# 2. **Histogram theo nhóm**
g = sns.FacetGrid(df, col='phanlopT1', col_order=labels, height=4, aspect=1)
g.map(sns.histplot, 'T1', kde=True, bins=15)
plt.suptitle("Histogram của T1 theo nhóm phanlopT1", y=1.05)
plt.show()

# 3. **QQ-Plot theo nhóm**
for group in labels:
    plt.figure(figsize=(6, 4))
    stats.probplot(df[df['phanlopT1'] == group]['T1'], dist="norm", plot=plt)
    plt.title(f"QQ-Plot của T1 cho nhóm {group}")
    plt.xlabel("Lý thuyết phân phối chuẩn")
    plt.ylabel("Phân phối thực tế")
    plt.show()
```

```python
# Câu 3: Hãy khảo sát tương quan giữa biến DH1 theo biến T1 

# Tính covariance và correlation
covariance = df['DH1'].cov(df['T1'])
correlation = df['DH1'].corr(df['T1'])
print("Covariance giữa DH1 và T1:", covariance)
print("Correlation giữa DH1 và T1:", correlation)

# Vẽ Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='T1', y='DH1', data=df)
plt.title("Scatter Plot giữa DH1 và T1")
plt.show()
```

```python
# Câu 4: Hãy khảo sát tương quan giữa biến DH1 theo biến T1 trên từng nhóm khu vực

# Vẽ Scatter Plot theo nhóm khu vực
g = sns.FacetGrid(df, col='KV', height=4, aspect=1)
g.map(sns.scatterplot, 'T1', 'DH1')
plt.suptitle("Scatter Plot giữa DH1 và T1 theo khu vực", y=1.05)
plt.show()
```

```python
# Câu 5: Khảo sát tương quan giữa các biến DH1, DH2, DH3

# Ma trận hiệp phương sai
cov_matrix = df[['DH1', 'DH2', 'DH3']].cov()
print("Ma trận hiệp phương sai:\n", cov_matrix)

# Ma trận tương quan
corr_matrix = df[['DH1', 'DH2', 'DH3']].corr()
print("Ma trận tương quan:\n", corr_matrix)

# Vẽ Pairplot
sns.pairplot(df[['DH1', 'DH2', 'DH3']])
plt.suptitle("Pairplot giữa DH1, DH2, DH3", y=1.05)
plt.show()
```
